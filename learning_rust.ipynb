## Hello, world

The code below would be in a file named `main.rs` and will simply print `"Hello, world!"`.

````rust
// main.rs
fn main() {
	println!("Hello, world!");
}
````

Typically to run this, we would use the following commands in the terminal:

```bash
$ rustc main.rs
$ ./main
```

And the `main.rs` file would run. However, in these notebooks, I'll simply run the function.

````rust
main();
````

The code is similar to python but seems to be sort of *mixed* with Java or C#. If we break down the code, the `fn main()` line defines a function named `main`. The `main` function is **always the first code that runs in every executable Rust program.** If we wanted to use parameters in our function, they would go inside of the parentheses `()` of the declaration, for example:

````rust
fn greet(name: &str) {
	println!("Hello, {name}!");
}

greet("Chris");
````

The body of the function exists wrapped in `{}`. **Rust requires curly brackets around all function bodies.**

Inside of the previous function, `println!("Hello, world!");` is the body. `println!` calls a Rust macro. If we were to call a function instead, we'd enter `println` without the `!`. Rust macros are *a way to write code that generates code to extend Rust syntax* (it'll be discussed later). Macros don't always follow the same rules as functions.

We end lines of code with a semicolon `;`, similar to Java.
